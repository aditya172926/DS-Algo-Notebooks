# Gradient Boosting
We have made predictions using the random forest method, which achieves better performance that a single decision tree simply bt averaging the predictions of many decision trees.<br>
We refer the random forest as ensemble method. By definition the ensemble methods combine the predictions of several models, eg several trees in case of random forest.<br>
Gradient Boosting is a method that goes through cycles to iteratively add models into an ensemble.<br>
It begins by initializing the ensemble with a single model, whose predictions can be preety naive. Even if its predictions are widely inaccurate, subsequent additions to the ensemble will addree those errors.<br><br>

Then we start the cycle:<br>
First we use the current ensemble to generate predictions for each observation in the dataset. To make prediction we add the predictions from all models in the ensemble.<br>
THese predictions are used to calculate a loss function like mean squared error.<br>
THen we use the loss function to fit a new model that will be added to ensemble. Specifically we determine model parameters so that adding this new model to the ensemble will reduce the loss.<br><br>

The Gradient in gradient boosting refers to the fact that we'll use gradient descent on the loss funtion to determine the parameters in this new model.
<br>
Gradient descent is a first order iterative optimization algorithm for finding a local minimum of a differentiable function.<br>
Finally, we add the new model to the ensemble and repeat!




```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
data = pd.read_csv('Datasets/melb_data.csv')

# Select subset of predictors
cols_to_use = ['Rooms', 'Distance', 'Landsize', 'BuildingArea', 'YearBuilt']
X = data[cols_to_use]

# Select target
y = data.Price

# Separate data into training and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

We will work with XGBoost library. It stands for Extrame gradient boosting which is an implementation of gradient boosting with several additional features focused on performance and speed.<br>
In the next call we import the scikit learn API for XGBoost (xgboost.XGBBRegressor). This allows us to build and fit a model just as we would in scikit-learn. As you'll see in the output, the XGBRegressor class has many tunable parameters.

```python
from xgboost import XGBRegressor

my_model = XGBRegressor()
my_model.fit(X_train, y_train)
```
