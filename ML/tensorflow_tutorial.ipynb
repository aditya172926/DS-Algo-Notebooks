# Introducing Tensors
To understand tensors well, it’s good to have some working knowledge of <b>linear algebra</b> and <b>vector calculus</b>. You already read in the introduction that tensors are implemented in TensorFlow as multidimensional data arrays, but some more introduction is maybe needed in order to completely grasp tensors and their use in machine learning.

## Plane Vectors
Before you go into plane vectors, it’s a good idea to shortly revise the concept of “vectors”; <b>Vectors are special types of matrices, which are rectangular arrays of numbers</b>. Because vectors are <b>ordered collections of numbers</b>, they are often seen as <b>column matrices</b>: they have just one column and a certain number of rows. In other terms, you could also consider vectors as scalar magnitudes that have been given a direction.
<br><br>
An example of a scalar is “5 meters” or “60 m/sec”, while a vector is, for example, “5 meters north” or “60 m/sec East”. The difference between these two is obviously that the vector has a <b>direction</b>. Nevertheless, these examples that you have seen up until now might seem far off from the vectors that you might encounter when you’re working with machine learning problems. This is normal; The length of a mathematical vector is a pure number: it is absolute. The direction, on the other hand, is relative: it is measured relative to some reference direction and has units of radians or degrees. You usually assume that the direction is positive and in counterclockwise rotation from the reference direction.



Visually, of course, you represent vectors as arrows, as you can see in the picture above. This means that you can consider vectors also as arrows that have direction and length. The direction is indicated by the arrow’s head, while the length is indicated by the length of the arrow.
<br><br>
So what about plane vectors then?
<br><br>

Plane vectors are the most <b>straightforward</b> setup of <b>tensors</b>. They are much like regular vectors as you have seen above, with the sole difference that they find themselves in a <b>vector space</b>. To understand this better, let’s start with an example: you have a vector that is 2 X 1. This means that the vector belongs to the set of real numbers that come paired two at a time. Or, stated differently, they are part of two-space. In such cases, you can represent vectors on the coordinate <b>(x,y)</b> plane with arrows or rays.
<br><br>
Working from this coordinate plane in a standard position where vectors have their endpoint at the origin (0,0), you can derive the x coordinate by looking at the first row of the vector, while you’ll find the y coordinate in the second row. Of course, this standard position doesn’t always need to be maintained: vectors can move parallel to themselves in the plane without experiencing changes.

<br><br>
Note that similarly, for vectors that are of size <b>3 X 1</b>, you talk about the three-space. You can represent the vector as a <b>three-dimensional</b> figure with arrows pointing to positions in the vectors pace: they are drawn on the standard <b>x, y and z axes</b>.

<br><br>
It’s nice to have these vectors and to represent them on the coordinate plane, but in essence, you have these vectors so that you can perform operations on them and one thing that can help you in doing this is by expressing your vectors as <b>bases</b> or <b>unit vectors</b>.
<br><br>
Unit vectors are vectors with a <b>magnitude of one</b>. You’ll often recognize the unit vector by a lowercase letter with a circumflex, or “hat”. Unit vectors will come in convenient if you want to <b>express</b> a <b>2-D</b> or <b>3-D vector</b> as a <b>sum</b> of <b>two or three orthogonal components</b>, such as the x− and y−axes, or the z−axis.
<br><br>
And when you are talking about expressing one vector, for example, as sums of components, you’ll see that you’re talking about component vectors, which are two or more vectors whose sum is that given vector.

## Tensors
Next to plane vectors, also <b>covectors</b> and <b>linear operators</b> are two other cases that all three together have one thing in common: they are <u>specific cases of tensors</u>. You still remember how a vector was characterized in the previous section as scalar magnitudes that have been given a direction. A tensor, then, is the <b>mathematical representation</b> of a <b>physical entity</b> that may be <u>characterized by magnitude and multiple directions</u>.
<br><br>
And, just like you represent a scalar with a single number and a <b>vector</b> with a <b>sequence</b> of <b>three numbers</b> in a <b>3-dimensional space</b>, for example, a tensor can be represented by an array of <b>3R numbers</b> in a 3-dimensional space.
<br><br>
The <b>“R”</b> in this notation represents the <b>rank</b> of the tensor: this means that in a <b>3-dimensional space</b>, a <b>second-rank tensor</b> can be represented by <u>3 to the power of 2 or 9 numbers</u>. In an <b>N-dimensional space</b>, <b>scalars</b> will still require only <b>one number</b>, while <b>vectors</b> will require <b>N numbers</b>, and <b>tensors</b> will require <b>N^R numbers</b>. This explains why you often hear that <b>scalars are tensors of rank 0</b>: since they have no direction, you can represent them with one number.
<br><br>
With this in mind, it’s relatively easy to recognize scalars, vectors, and tensors and to set them apart: <b>scalars</b> can be represented by a <b>single number</b>, <b>vectors</b> by an ordered <b>set of numbers</b>, and <b>tensors</b> by an <b>array of numbers</b>.
<br><br>
What makes tensors so <b>unique</b> is the <b>combination</b> of <b>components and basis vectors</b>: basis vectors transform one way between reference frames and the components transform in just such a way as to keep the combination between components and basis vectors the same.

Let’s get started with importing the data into your workspace. Let’s start with the lines of code that appear below the User-Defined Function (UDF) load_data():

1. First, set your ROOT_PATH. This path is the one where you have made the directory with your training and test data.

Now, the load_data() function itself starts off by gathering all the subdirectories that are present in the train_data_directory; It does so with the help of list comprehension, which is quite a natural way of constructing lists - it basically says that, if you find something in the train_data_directory, you’ll double check whether this is a directory, and if it is one, you’ll add it to your list. 

Remember that each subdirectory represents a label.

Next, you have to loop through the subdirectories. You first initialize two lists, labels and images. Next, you gather the paths of the subdirectories and the file names of the images that are stored in these subdirectories. After, you can collect the data in the two lists with the help of the append() function.

```python
import os
from skimage import io
def load_data(data_directory):
    directories = [d for d in os.listdir(data_directory) if os.path.isdir(os.path.join(data_directory, d))]
    labels = []
    images = []
    
    for d in directories:
        label_directory = os.path.join(data_directory, d)
        file_names = [os.path.join(label_directory, f)
                     for f in os.listdir(label_directory)
                     if f.endswith(".ppm")]
        
        for f in file_names:
            images.append(io.imread(f))
            labels.append(int(d))
    return images, labels
        

ROOT_PATH = '/Users/aditya/Desktop/Kaggle-Notebooks/ML/data'
train_data_directory = os.path.join(ROOT_PATH, 'Training')
test_data_directory = os.path.join(ROOT_PATH, '/Testing')

images, labels = load_data(train_data_directory)

    
```

## Modeling the Neural Network
If you haven’t done so already, import tensorflow into your workspace under the conventional alias tf. Then, you can initialize the Graph with the help of <b>Graph()</b>. You use this function to <b>define the computation</b>. Note that with the <u>Graph, you don’t compute anything</u>, because it doesn’t hold any values. It just defines the operations that you want to be running later.

In this case, you set up a <b>default context</b> with the help of <b>as_default()</b>, which returns a context manager that makes this specific Graph the default graph. You use this method if you want to <b>create multiple graphs</b> in the same process: with this function, you have a global default graph to which all operations will be added if you don’t explicitly create a new graph.

Next, you’re ready to add operations to your graph. As you might remember from working with Keras, you build up your <b>model</b>, and then in compiling it, you <b>define a loss function</b>, an <b>optimizer</b>, and a <b>metric</b>. This now all happens in one step when you work with TensorFlow:

1. First, you define <b>placeholders for inputs and labels</b> because you won’t put in the “real” data yet. Remember that placeholders are values that are <b>unassigned</b> and that will be initialized by the session when you run it. So when you finally run the session, these placeholders will get the values of your dataset that you pass in the run() function!
2. Then, you build up the network. You first start by <b>flattening the input</b> with the help of the <b>flatten()</b> function, which will give you an array of shape <b>[None, 784]</b> instead of the [None, 28, 28], which is the shape of your grayscale images.
3. After you have flattened the input, you construct a <b>fully connected layer that generates logits of size [None, 62]</b>. Logits is the function operates on the unscaled output of previous layers, and that uses the relative scale to understand the units is linear.
4. With the multi-layer perceptron built out, you can define the loss function. The choice for a loss function depends on the task that you have at hand: in this case, you make use of:
`sparse_softmax_cross_entropy_with_logits()`
This <u>computes sparse softmax cross entropy</u> between logits and labels. In other words, it measures the <b>probability error</b> in <u>discrete classification tasks</u> in which the <u>classes are mutually exclusive</u>. This means that <b>each entry is in exactly one class</b>. Here, a traffic sign can only have one single label. Remember that, while <b>regression</b> is used to <b>predict continuous values</b>, <b>classification</b> is used to <b>predict discrete values or classes</b> of data points. You wrap this function with <b>reduce_mean()</b>, which computes the mean of elements across dimensions of a tensor.
5. You also want to define a training optimizer; Some of the most popular optimization algorithms used are the Stochastic Gradient Descent (SGD), ADAM and RMSprop. Depending on whichever algorithm you choose, you’ll need to tune certain parameters, such as learning rate or momentum. In this case, you pick the ADAM optimizer, for which you define the learning rate at 0.001.


```python
import tensorflow as tf

# initialize the placeholders
x = tf.placeholder(dtype = tf.float32, shape = [None, 28, 28])
y = tf.placeholder(dtype = tf.int32, shape = [None])

# flatten the input data
images_flat = tf.contrib.layers.flatten(x)

# fully connected layer
logits = tf.contrib.layers.fully_connected(images_flat, 62, tf.nn.relu)

# defina a loss function
loss = tf.reduce_mean(tf.nn.sparse_softmax_cross_entropy_with_logits(labels = y, logits = logits))

# define an optimizer
train_op = tf.train.AdamOptimizer(learning_rate=0.001).minimize(loss)

# convert logits ito label indexes
correct_pred = tf.argmax(logits, 1)

# defina an accuracy metric
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```
