## Changes over time
Statistical analysis methods are often based on changes over time and not the absolute values themselves. There are multiple options to calculate the changes in a time series over time, including <b>absolute differences, percentage changes, and logarithmic
(log) returns</b>.

First the absolute differences for which pandas gives special method

```python
# absolute differences
data.diff().head()
```

```python
data.diff().mean()
```

<b>diff()</b> provides the absolute changes between 2 index values.

From a statistics point of view, absolute changes are not optimal because they are dependent on the scale of the time series data itself. Therefore, percentage changes
are usually preferred. The following code derives the percentage changes or percentage returns (also: simple returns) in a financial context and visualizes their mean values per column.

```python
data.pct_change().round(3).head()
```

```python
data.pct_change().mean().plot(kind='bar', figsize=(10, 6))
```

As an alternative to percentage returns, log returns can be used. In some scenarios, they are easier to handle and therefore often preferred in a financial context. Shows the cumulative log returns for the single financial time series. This type of plot leads to some form of normalization.

```python
#Calculates the returns in vectorized fashion
rets = np.log(data / data.shift(1))
# subset of results
rets.head().round(3)
```

```python
rets.cumsum().apply(np.exp).plot(figsize=(10, 6))
```

to one-minute intervals or a time series with daily observations is resampled to one with weekly or monthly observations (as shown in Figure 8-4)

## Avoiding Foresight Bias
When resampling, pandas takes by default in many cases the left label (or index value) of the interval. To be financially consistent, make sure to use the <b>right label</b> (index value) and in general the last available data point in the interval. Otherwise, a foresight bias might sneak into the financial analysis.

Foresight biasor, in its strongest form, perfect foresightmeans that at some point in the financial analysis, data is used that only becomes available at a later point. The result might be too good results, for example, when backtesting a trading strategy.

# Rolling Statistics
It is financial tradition to work with <i>rolling statistics</i>, often called <b>financial indicators</b>. It is a basic tool for financial chartists and technical traders. This section works with a single financial time series only.

```python
sym = 'MMM'
data = pd.DataFrame(data[sym]).dropna()
data.tail()
```

## Overview
It is straightforward to derive standard rolling statistics with pandas.

To derive more specialized financial indicators, additional packages are generally needed (see, for instance, the financial plots with Cufflinks in Interactive 2D Plot ting). Custom ones can also easily be applied via the <b>apply() method</b>.

```python
# Defined a window, that is the number of index values to include
window=20
# Calculates the rolling minimum value
data['min'] = data[sym].rolling(window=window).min()
# Calculates the rolling mean value
data['mean'] = data[sym].rolling(window=window).mean()
# Calculates the rolling standard deviation
data['std'] = data[sym].rolling(window=window).std()
# Calculates the rolling median value
data['median'] = data[sym].rolling(window=window).median()
# Calculates the rolling maximum value
data['max'] = data[sym].rolling(window=window).max()
# Calculates the exponentially weighted moving average, with decay in terms of halflife of 0.5
data['ewma'] = data[sym].ewm(halflife=0.5, min_periods=window).mean()
```

```python
data.dropna().head()
```

```python
# plots 3 rolling statistics for the final 200 data rows
ax = data[['min', 'mean', 'max']].iloc[-200:].plot(figsize=(10, 6),
                                                  style=['g--', 'r--', 'g--'],
                                                  lw=0.8)
# Adds the original time series data to the plot
data[sym].iloc[-200:].plot(ax=ax, lw=2.0)
```

# A Technical Analysis Example
Rolling statistics are a major tool in the so called <b>technical analysis</b> of stocks, as compared to the fundamental analysis which focuses for instance on financial reports and the strategic positions of the company whose stock is being analysed.

A decades-old trading strategy based on technical analysis is using <b>two simple moving averages (SMAs)</b>. The idea is that the trader should go <b>long on a stock</b> (or financial instrument in general) when the <b>shorter-term SMA is above the longer-term SMA</b> and should go <b>short</b> when the <b>opposite holds true</b>. The concepts can be made precise with pandas and the capabilities of the DataFrame object.

Rolling statistics are generally only calculated when there is enough data given the window parameter specification. As Figure 8-6 shows, the SMA time series only start at the day for which there is enough data given the specific parameterization.

```python
# calculates teh values for shorter SMA
data['SMA1']=data[sym].rolling(window=42).mean()
# calculates the values for larger SMA
data['SMA2']=data[sym].rolling(window=252).mean()
data[[sym, 'SMA1', 'SMA2']].tail()
```

```python
data[[sym, 'SMA1', 'SMA2']].plot(figsize=(10, 6))
```

In this context, the SMAs are only a means to an end. They are used to derive posi tions to implement a trading strategy. Figure 8-7 visualizes a long position by a value of 1 and a short position by a value of -1. The change in the position is triggered (vis ually) by a crossover of the two lines representing the SMA time series.

```python
data.dropna(inplace=True)
data['positions'] = np.where(data['SMA1'] > data['SMA2'], 1, -1)
ax = data[[sym, 'SMA1', 'SMA2', 'positions']].plot(figsize=(10, 6),
                                                  secondary_y='positions')
ax.get_legend().set_bbox_to_anchor((0.25, 0.85))
```
